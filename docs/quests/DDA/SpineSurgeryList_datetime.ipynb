## 시작

```python
get_data = pd.read_csv('SpineSurgeryList.csv')
get_data
```

```python
get_data.columns
```

```python
get_data_long = get_data[['환자ID','입원기간']].sort_values('입원기간')
get_data_long
```

```python
in_date = pd.to_datetime(get_data['입원일자'], format='%Y%m%d')
out_date = pd.to_datetime(get_data['퇴원일자'], format='%Y%m%d')
get_data['입원기간_second'] = out_date - in_date
get_data['입원기간_second'] = get_data['입원기간_second'].dt.days
```

```python
get_data_long_second = get_data[['환자ID','입원기간_second']].sort_values('입원기간_second')
get_data_long_second.info()
```

```python
sns.barplot(data = get_data_long_second[-5:], x='환자ID', y='입원기간_second')
```

```python
sns.barplot(data = get_data_long[-5:], x='환자ID', y='입원기간')
```

```python
get_data_when = get_data[['환자ID','수술일자']]
get_data_when.info()
```

```python
get_data_when['수술일자_datetime'] = pd.to_datetime(get_data_when['수술일자'], format='%Y%m%d')
get_data_when['수술일자_datetime_month'] = get_data_when['수술일자_datetime'].dt.month
get_data_when['수술일자_datetime_day'] = get_data_when['수술일자_datetime'].dt.day
get_data_when['수술일자_datetime_week'] = get_data_when['수술일자_datetime'].dt.isocalendar().week
get_data_when_visual = get_data_when[['수술일자_datetime_month','수술일자_datetime_day','수술일자_datetime_week']].to_frame()
get_data_when_visual.value_counts()
```

```python
sns.countplot(data=get_data_when_visual, y='수술일자_datetime_month')
```

```python
sns.countplot(data=get_data_when_visual, y='수술일자_datetime_day')
```

```python
sns.countplot(data=get_data_when_visual, y='수술일자_datetime_week')
```
