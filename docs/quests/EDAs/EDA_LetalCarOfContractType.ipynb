### 결측치 처리

```python
get_data.isnull().sum()
```

- 마일리지와 나이가 같은 숫자이므로 개인정보를 저장하고 싶지 않고 마일리지를 가입하고 싶지않은 해당 결측치가 대부분 1회성 고객일 것으로 예상
- 1회성 고객임을 알기위해 나이 결측치로 추출 후 계약 내용을 검토



```python
check_ontiem = get_data[get_data['age'].isna()]
check_ontiem.isnull().sum()
```

- 모든 나이와 마일리지 결측치가 서로 그룹화되어있는것을 확인

```python
check_ontiem['type_of_contract'].value_counts()
```

- 장기고객인 멤버십보다 1회성인 렌탈이 매우 많은 것을 확인
- 멤버십 고객의 충성도를 확인하기 위해 계약기간과 지불방법을 검토

```python
check_ontiem.query('type_of_contract == "멤버십"')['Term'].value_counts()
```

- 해당되는 멤버십 고객들이 단기 기간 계약을 유지하고 있는 상황임을 확인

```python
check_ontiem.query('type_of_contract == "멤버십"')['payment_type'].value_counts()
```

- 자동 이체(CMS)보다 무통장입금(1회성)이 월등히 많음 => 아직까지 이용에 자동이체를 할 만큼의 신뢰도 부족으로 판단
- 결측치 그룹을 따로 분석할 필요성 있음 -> 1회성 고객을 충성고객으로 전환하기 위한 데이터로 이용 등
