## 시작

```python
get_data = pd.read_csv('TitanicFromDisaster_train.csv')
get_data
```

### 문제 1. Age 열에서 결측치를 평균 나이로 채우기

```python
cal_age = get_data['Age']
cal_age
```

```python
cal_age.dropna(inplace=True)
cal_age
```

```python
mean = round(cal_age.mean())
mean
```

```python
get_data['Age'] = get_data['Age'].replace(np.nan,mean)
get_data
```

### 문제 2. cabin 열에 결측치가 있는 행을 찾고, 해당 행들의 평균 Fare 계산하기

```python
get_data['Cabin_isnull'] = get_data['Cabin'].isnull()
get_data
```

```python
get_data.query('Cabin_isnull == True')['Fare'].mean()
```

### 문제 3. Embarked 열에서 결측치가 있는 행을 찾고, 가장 많이 등장하는 항구로 채우기

```python
max_key = get_data['Embarked'].value_counts().index[0]
max_key
```

```python
get_data['Embarked'] = get_data['Embarked'].replace(np.nan,max_key)
get_data['Embarked']
```

```python
get_data['Embarked'].isnull().sum()
```

### 문제 4. Fare 열에서 평균보다 3표준편차 이상 높은 이상치를 가진 행들을 찾기

```python
fare_mean = get_data['Fare'].mean()
fare_std = get_data['Fare'].std()
get_data[get_data['Fare'] > fare_mean + fare_std*3]
```

### 문제 5. SibSp (형제 자매/배우자수)가 5 이상인 이상치를 가진 행들을 찾기

```python
get_data.query('SibSp >= 5')
```

### 문제 6. Parch (부모/자녀 수) 열에서 0이 아닌 값을 가진 행들의 평균 Age 계산하기.

```python
get_data.query('Parch != 0')['Age'].mean()
```

### 문제 7. Fare 가 0인 행들을 찾고, 이를 평균 Fare로 채우기

```python
get_data.query('Fare == 0')
```

```python
get_data['Fare'] = get_data['Fare'].replace(0.0,fare_mean)
get_data.query('Fare == 0')
```

### 문제 8. Pclass (티켓 등급)가 1등급인 승객들 중 Cabin 열의 결측치 비율 계산하기

```python
count_total = get_data.query('Pclass == 1')['Cabin'].value_counts().sum()
count_total
```

```python
get_data.query('Pclass == 1')['Cabin'].unique()
```

```python
# count_nan = get_data.query(f'Pclass == 1 & Cabin == {np.nan}')['Cabin'].value_counts().sum()
count_nan = get_data.query('Pclass == 1 & Cabin == "nan"')['Cabin'].value_counts().sum()
count_nan
```

```python
answer = (count_nan / count_total)*100
answer
```
