```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
```

```python
get_data = pd.read_csv('ShoppingMallDeliveryWithDate.csv')
```

```python
get_data
```

```python
orderpath = get_data['주문경로'].value_counts()
orderpath_frame_top = orderpath.iloc[:3].to_frame()
orderpath_frame_top = orderpath_frame_top.rename(columns={'주문경로':'회수'})
orderpath_frame_top.index.name = '주문경로'
orderpath_frame_top
```

```python
sns.barplot(data=orderpath_frame_top, x='회수', y='주문경로')
```

```python
orderpath = get_data['주문경로'].value_counts()
orderpath_frame_low = orderpath.iloc[-3:].to_frame()
orderpath_frame_low = orderpath_frame_low.rename(columns={'주문경로':'회수'})
orderpath_frame_low.index.name = '주문경로'
orderpath_frame_low
```

```python
sns.barplot(data=orderpath_frame_low, x='회수', y='주문경로')
```
